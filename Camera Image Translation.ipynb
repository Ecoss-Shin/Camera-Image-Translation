# 카메라 이미지 번역

```python
from tkinter import *
from PIL import ImageTk, Image # Pillow
import cv2
import numpy as np
import pytesseract
import googletrans

# GUI 설계
win = Tk() # 인스턴스 생성

win.title("Ecoss_Solution") # 제목 표시줄 추가
win.option_add("*Font", "맑은고딕 25") # 전체 폰트

#win.resizable(False, False) # x축, y축 크기 조정 비활성화

# 라벨 추가
lbl = Label(win, text="카메라 이미지 번역")
lbl.grid(row=0, column=0) # 라벨 행, 열 배치

# 프레임 추가
frm = Frame(win, bg="white", width=720, height=480) # 프레임 너비, 높이 설정
frm.grid(row=1, column=0) # 격자 행, 열 배치

# 라벨 추가
lbl1 = Label(frm)
lbl1.grid()

cap = cv2.VideoCapture(1) # VideoCapture 객체 정의

def video_play():
   ret, frame = cap.read()
   frame = cv2.flip(frame, -1) # 상하반전
   frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

   img = Image.fromarray(frame)
   imgtk = ImageTk.PhotoImage(image=img)

   lbl1.imgtk = imgtk
   lbl1.configure(image=imgtk)
   lbl1.after(10, video_play)

video_play()

# 버튼 추가
btn = Button(win, text = "번역") # 버튼 생성
btn.grid(row=2, column=0) # 라벨 행, 열 배치
btn.config(width = 5, height = 1) # 버튼 크기

def alert():
    
    ret, frame = cap.read()
    frame = cv2.flip(frame, -1) # 상하반전
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
    text = pytesseract.image_to_string(gray, lang = "eng")

    print(text)
    
    translator = googletrans.Translator()

    result = translator.translate(text, dest='ko')

    result = result.text
    
    print("-----------------------------------\n")
    print(result)

btn.config(command = alert) # 버튼 기능

win.mainloop() #GUI 시작
```
